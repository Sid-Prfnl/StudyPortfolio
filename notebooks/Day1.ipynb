```python
import pandas as pd
import matplotlib.pyplot as plt

# Display settings
pd.set_option('display.max_columns', None)

print("Setup successful!")
```

```python
df = pd.read_csv("E:\Conda projects/data/Dirk.csv")

df.head()
df.info()
df.describe()
df.sort_values(by="3P%", ascending=False)

Best3P = (
    df.groupby("Season")["3P%"]
      .sum()
      .sort_values(ascending=False)
      .head(8)
)

Best3P.plot(kind="bar", title="Best 3Pt Shooting seasons")
plt.show()

```
